```python
Dealing with the null values
```

```python
 #linear interpolation for 'SuicideCount' as we assume that missing counts can be predicted based on trends in the data
df['SuicideCount'].interpolate(method='linear', inplace=True)
```

```python
#For 'CauseSpecificDeathPercentage' and 'DeathRatePer100K', we fill missing values with the mean
# This assumes that the missing data can be reasonably estimated from the average
df['CauseSpecificDeathPercentage'].fillna(df['CauseSpecificDeathPercentage'].mean(), inplace=True)
df['DeathRatePer100K'].fillna(df['DeathRatePer100K'].mean(), inplace=True)
```

```python
# Handle missing values in 'Population' using interpolation, as populations usually grow or change gradually over time
df['Population'].interpolate(method='linear', inplace=True)
```

```python
# Economic indicators in columns like 'GDP', 'GDPPerCapita', 'GrossNationalIncome') we fill them using either the mean or interpolation
# We assume smooth changes over time for interpolation or use the average for missing values
df['GDP'].fillna(df['GDP'].mean(), inplace=True)
df['GDPPerCapita'].interpolate(method='linear', inplace=True)
df['GrossNationalIncome'].fillna(df['GrossNationalIncome'].mean(), inplace=True)
df['GNIPerCapita'].interpolate(method='linear', inplace=True)
df['InflationRate'].fillna(df['InflationRate'].mean(), inplace=True)
df['EmploymentPopulationRatio'].fillna(df['EmploymentPopulationRatio'].mean(), inplace=True)
```

```python
#checking for any remaining null values
df.isnull().sum()
```

VISUALIZATION

```python
#Top ten countries with the highest suicide rates
#grouping by countryname and sum of the suicidecount
suicide_counts=df.groupby('CountryName')['SuicideCount'].sum().sort_values(ascending=False)

#plot for the top 10 countries with highest suicide counts
suicide_counts.head(10).plot(kind='bar',figsize=(10,6))
plt.title('Top 10 Countries By Total Suicide Counts')
plt.xlabel('Country')
plt.ylabel('Total Suicide Count')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

```python
#Group by Year and calculate total suicidecount for each year
suiciderates_by_year = df.groupby('Year')['SuicideCount'].sum()

#plotting the trend over time
suiciderates_by_year.plot(kind='line',figsize=(10,6))
plt.title('Global Suicide Counts Over Time')
plt.xlabel('Year')
plt.ylabel('Total Count of Suicide Counts Over Time')
plt.xticks(rotation=45)
plt.grid(True)
plt.show
```

```python
#Histogram for DeathRateper100k
df['DeathRatePer100K'].plot(kind='hist',bins=20,edgecolor='red')
plt.title('Distribution of suicide rates per 100k population')
plt.xlabel('Suicide Rate per 100k')
plt.ylabel('Frequency')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

```python
#scatterplot of GDDPerCapita vs DeathRateper100k
plt.figure(figsize=(10,6))
plt.scatter(df['GDPPerCapita'], df['DeathRatePer100K'], alpha=0.6)
plt.title('Suicide Rate vs GDP Per Capita')
plt.xlabel('GDP Per Capita')
plt.ylabel('Sucide Rate per 100k Population')
plt.grid(True)
plt.show
```

```python
#Box plot for suicide rates by sex
plt.figure(figsize=(10,6))
df.boxplot(column='DeathRatePer100K', by='Sex')
plt.title('Distribution of Suicide Rates Per Geneder')
plt.suptitle('')
plt.xlabel('Gender')
plt.ylabel('Suicide Rate Per 100k Population')
plt.show()
```

```python
#Calculating the correlation matrix
corr_matrix=df.select_dtypes(include='number').corr()
#plotting the heatmap
plt.figure(figsize=(12,8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('correlation Heatmap for Key variables')
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.show()

```

```python
# Group by 'RegionName' and 'Sex', summing the 'SuicideCount'
suicide_by_region_gender = df.groupby(['RegionName', 'Sex'])['SuicideCount'].sum().unstack()

# Plot a stacked bar chart
suicide_by_region_gender.plot(kind='bar', stacked=True, figsize=(12, 8))
plt.title('Suicide Counts by Region and Gender')
plt.xlabel('Region')
plt.ylabel('Total Suicide Count')
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
#Proportion of suicides by Age Group
Suicide_by_age_group=df.groupby('AgeGroup')['SuicideCount'].sum()
Suicide_by_age_group.plot(kind='pie', autopct='%1.1f%%', figsize=(8,8), startangle=90)
plt.title('Proportion of Suicides By Age Group')
plt.ylabel('')
plt.show()
```
